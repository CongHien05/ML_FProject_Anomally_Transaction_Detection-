Nạp dữ liệu vào bộ nhớ

```python
import pandas as pd

print("Đang nạp dữ liệu từ ổ cứng...")
# Nhớ điều chỉnh lại đường dẫn cho đúng với thư mục máy bạn
X_train = pd.read_csv('../data/X_train_final.csv')
y_train = pd.read_csv('../data/y_train.csv')
X_test = pd.read_csv('../data/X_test_final.csv')
y_test = pd.read_csv('../data/y_test.csv')

# Chuyển y thành mảng 1 chiều (1D array) để thư viện scikit-learn không báo lỗi
y_train = y_train.values.ravel()
y_test = y_test.values.ravel()

print(f"Kích thước tập Train (để học): {X_train.shape}")
print(f"Kích thước tập Test (để thi): {X_test.shape}")
```

Huấn luyện Rừng ngẫu nhiên (Training)
Vì dữ liệu thật rất lớn, chúng ta sẽ thêm tham số n_jobs=-1 để kích hoạt toàn bộ nhân CPU của máy tính, giúp quá trình huấn luyện diễn ra nhanh nhất có thể.

```python
from sklearn.ensemble import RandomForestClassifier
import time

# Khởi tạo cỗ máy
rf_model = RandomForestClassifier(
    n_estimators=100, 
    random_state=42, 
    n_jobs=-1  # Sử dụng tối đa sức mạnh phần cứng
)

print("Bắt đầu quá trình huấn luyện... (Có thể mất vài phút tùy cấu hình máy)")
start_time = time.time()

# Ép cỗ máy học trên tập Train
rf_model.fit(X_train, y_train)

end_time = time.time()
print(f"Huấn luyện thành công! Tổng thời gian: {end_time - start_time:.2f} giây.")
```

Tổ chức "Thi sát hạch" và chấm điểm
Đây là khoảnh khắc quyết định xem các đặc trưng toán học và bộ dữ liệu SMOTE của nhóm bạn có thực sự hiệu quả hay không.

```python
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# Bắt mô hình làm bài thi
print("Đang dự đoán trên tập Test...")
y_pred = rf_model.predict(X_test)

# Chấm điểm
print("\n=== BÁO CÁO PHÂN LOẠI (CLASSIFICATION REPORT) ===")
print(classification_report(y_test, y_pred, target_names=['An Toàn (0)', 'Gian Lận (1)']))

# Vẽ Ma trận nhầm lẫn
cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['An Toàn', 'Gian Lận'])

fig, ax = plt.subplots(figsize=(6, 6))
disp.plot(cmap='Blues', values_format='d', ax=ax)
plt.title('Ma trận Nhầm lẫn trên Dữ liệu Thật')
plt.grid(False)
plt.show()
```
